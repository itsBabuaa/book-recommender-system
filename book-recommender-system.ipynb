## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean(numeric_only = True).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on= 'Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

## Collaborative Filtering Based Recommender System

```python
# taking user who rated more than 200 book
temp_rated_over_200 = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
rated_over_200 = temp_rated_over_200[temp_rated_over_200].index #taking index of the users
```

```python
rated_over_200
```

```python
#filtering the ratings
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(rated_over_200)]
```

```python
filtered_rating
```

```python
# taking books with at least 50 rating only
rated_50 = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = rated_50[rated_50].index
```

```python
famous_book
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns= 'User-ID', values= 'Book-Rating')
```

```python
pt.fillna(0, inplace= True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #index fetch
    #index = np.where(pt.index==book_name)[0][0]
    index = np.where(pt.index==book_name) [0]
    index = index[0]
    distance = similarity_scores[index]
    similar_items = sorted(list(enumerate(distance)),key= lambda x:x[1], reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
    #return suggestions
```

```python
recommend('Life of Pi')
```

```python
import pickle
from sklearn.metrics.pairwise import cosine_similarity

# Load your merged ratings dataset (replace with your file if needed)
# Example: final_ratings = pd.read_csv('final_ratings.csv')
# It should contain: User-ID, Book-Title, Book-Rating

# Create the pivot table
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)

# Compute similarity matrix
similarity_scores = cosine_similarity(pt)

# Save pivot table
with open('pt.pkl', 'wb') as f:
    pickle.dump(pt, f)

# Save similarity scores
with open('similarity_scores.pkl', 'wb') as f:
    pickle.dump(similarity_scores, f)

# Popular books dataframe for home page
popular_df = final_ratings.groupby('Book-Title').agg({
    'Book-Rating': ['count', 'mean']
}).reset_index()

popular_df.columns = ['Book-Title', 'num_ratings', 'avg_rating']

# Load the books dataset (should have Book-Title, Book-Author, Image-URL-M)
books = pd.read_csv('books.csv')

popular_df = popular_df.merge(books[['Book-Title', 'Book-Author', 'Image-URL-M']], on='Book-Title').drop_duplicates('Book-Title')

# Save popular_df
with open('popular.pkl', 'wb') as f:
    pickle.dump(popular_df, f)

# Save books df
with open('books.pkl', 'wb') as f:
    pickle.dump(books, f)
```
